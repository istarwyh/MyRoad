```python
import matplotlib.pyplot as plt
import logging
from mpl_toolkits.mplot3d import Axes3D

logging.basicConfig(level=logging.INFO)

def setup_axes():
    """"""
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    # 
    ax.spines['left'].set_position('zero')
    ax.spines['bottom'].set_position('zero')

    # 
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')

    # 
    ax.quiver(0, 0, 0, 1, 0, 0, length=1, normalize=True)
    ax.quiver(0, 0, 0, 0, 1, 0, length=1, normalize=True)
    ax.quiver(0, 0, 0, 0, 0, 1, length=1, normalize=True)

    logging.info("Axes set up complete.")
    return ax

def plot_points(ax, x, y, z, labels):
    """"""
    ax.scatter(x, y, z)
    for i in range(len(x)):
        ax.text(x[i], y[i], z[i], labels[i])
    logging.info("Points plotted.")

def finalize_plot(ax):
    """"""
    ax.set_xlim([0, 1])
    ax.set_ylim([0, 1])
    ax.set_zlim([0, 1])
    plt.show()
    logging.info("Plot displayed.")

# 
ax = setup_axes()
x = [0, 1, 1, 1]
y = [0, 0, 1, 1]
z = [0, 1, 1, 0]
labels = ['King', 'Queen', 'Man', 'Woman']
plot_points(ax, x, y, z, labels)
finalize_plot(ax)
```
